```python
import os
import re

root_folder = r"C:\Repos\Snowtide\City_of_SanDiego\Municode\ExtractMuniCode\municipal_code_pdfs"
total_pdfs = 0

for subfolder in sorted(os.listdir(root_folder)):
    subfolder_path = os.path.join(root_folder, subfolder)
    if os.path.isdir(subfolder_path):
        pdf_count = sum(1 for f in os.listdir(subfolder_path) if f.lower().endswith(".pdf"))
        total_pdfs += pdf_count
        # Extract Chapter_XX and optional name
        match = re.match(r"(Chapter_\d+)(_.*)?", subfolder, re.IGNORECASE)
        chapter_label = match.group(0) if match else subfolder
        print(f"{chapter_label}: {pdf_count} PDFs")

print(f"\nTotal PDFs across all chapters: {total_pdfs}")
```
